```python
import sys
from utilities import get_data,display_metrics
from sklearn.svm import LinearSVC as SVC
from sklearn.metrics import accuracy_score
a= []
for i in range (1,101):
    clf = SVC(multi_class= 'ovr', max_iter=i)
    x_train, x_test, y_train, y_test = get_data()
    #print ('------------- I am learning, yay! -------------')
    clf.fit(x_train, y_train)
    #print ('------------- I finished learning :) -------------')
    y_pred = clf.predict(x_test)
    b = float(accuracy_score(y_pred=y_pred, y_true=y_test))
    a.append(b)
print(a)    
    
```

```python
import matplotlib.pyplot as plt
plt.plot(a)
plt.show()
```

```python
# module 5 exercises

import numpy as np
import matplotlib.pyplot as plt


# 10
def lagrange_poly(k, N, t):
    res = 1
    for i in np.arange(-N, N+1, 1):
        if i != k:
            res *= (t - i)/(k - i)
    return res

def lag_interp(s, k, N, t):
    res = 0
    for ss, kk in zip(s, k):
        res += ss*lagrange_poly(kk, 1, t)
    return res

if __name__ == "__main__":
    s = np.array([4, -2, -3])
    k = np.array([-1, 0, 1])
    res = lag_interp(s, k, 1, 1/2)
    print(res)

    x = np.linspace(-1, 1, 50)
    plt.plot(x, lag_interp(s, k, 1, x))
    plt.show()
```
